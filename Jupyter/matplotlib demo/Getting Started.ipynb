```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("First Plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("First Plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
x = np.arange(1,5)
y = x**3
plt.plot([1,2,3,4],[1,4,9,16],"go", x, y, 'r^')
plt.title("First Plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("1st subplot")

plt.subplot(1,2,2)
plt.plot(x, y, 'r^')
plt.title("1st subplot")

plt.suptitle("My sub-plöots")
plt.show()
```

```python
# Bar graph
divisions = ["Div-A","Div-B","Div-C","Div-D","Div-E"]
division_average_marks =[70,82,73,65,68]

plt.bar(divisions, division_average_marks, color = 'green')
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
# Pie chart
firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E" ]
market_share = [20, 25, 15, 2, 28]
Explode = [0,0.1,0,0,0]
plt.pie(market_share, explode=Explode, labels=firms,shadow=True, startangle=45)
plt.axis('equal')
plt.legend(title="List of Firms")
plt.show()
```

```python
# Histogram
x = np.random.randn(1000)

plt.title("Histogram")
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.hist(x,10)
plt.show()
```
